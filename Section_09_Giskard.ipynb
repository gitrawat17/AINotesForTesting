````python
# Load the dataset
df = pd.read_csv('/content/ecommerceDataset.csv', header=None, names=['Label', 'Text'])
df = df.sample(1000)
````

````python
# Load pre-trained model tokenizer
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')


# Load pre-trained model and adjust for 4 labels
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=4)
model.to(device)
````

Define a custom PyTorch data pipeline to make predictions on a dataset using a fine-tuned Hugging Face BERT model. Here's an explanation of the key components:

1. **`PredictDataset` Class**:  
   This is a custom dataset class that prepares text data for predictions. It takes the tokenized input and returns it as a PyTorch tensor, which is required for processing with the model. The `__getitem__()` method ensures that each tokenized text is returned as a dictionary of tensors (`input_ids` and `attention_mask`), and `__len__()` provides the dataset's size.

2. **`custom_predict(df)` Function**:  
   This function takes a DataFrame (`df`) and:
   - Extracts the texts from the `Text` column of the DataFrame.
   - Tokenizes the texts using the pre-loaded BERT tokenizer (`tokenizer()`).
   - Wraps the tokenized inputs in the `PredictDataset` class and creates a `DataLoader` for batching (with a batch size of 16, without shuffling).
   - Iterates over each batch, moves the input tensors to the appropriate device (CPU or GPU), and uses the model to make predictions. The function calculates the softmax probabilities for each class.
   - The predicted probabilities are collected and returned as a list.

This setup allows for efficient batching and prediction over large datasets, ensuring that inputs are tokenized, moved to the GPU (if available), and processed by the BERT model to generate class probabilities.

````python
class PredictDataset(torch.utils.data.Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __getitem__(self, idx):
        return {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}

    def __len__(self):
        return len(self.encodings['input_ids'])

def custom_predict(df):
    texts = df['Text'].tolist()
    encodings = tokenizer(texts, truncation=True, padding=True, max_length=512)  # Adjust max length if necessary

    predict_dataset = PredictDataset(encodings)
    loader = DataLoader(predict_dataset, batch_size=16, shuffle=False)
    probabilities = []

    for batch in loader:
        with torch.no_grad():
            input_batch = batch['input_ids'].to(device)  # Move input to GPU if available
            mask_batch = batch['attention_mask'].to(device)  # Move attention mask to GPU if available
            output = model(input_ids=input_batch, attention_mask=mask_batch)
            logits = output.logits
            probs = F.softmax(logits, dim=1)
            probabilities.extend(probs.cpu().tolist())  # Move back to CPU before converting to list

    return probabilities
````

**Label Encoding:**

The column Label in the dataset is mapped to numerical values using a dictionary. Categories such as Electronics, Household, Books, and Clothing & Accessories are assigned values 0, 1, 2, and 3, respectively. The resulting column LabelNum contains these numerical representations, which are necessary for training a classification model.

**Train-validation Split:**

The dataset is split into training and validation sets using an 80-20 split (test_size=0.8). The train_test_split() function ensures that 80% of the data goes to the validation set, while 20% is used for training. The random_state=42 ensures reproducibility of the split.

**Tokenization:**

Both the training and validation text data are tokenized using the pre-trained BERT tokenizer. The tokenizer() function converts the text into input tokens (numerical representations) and pads or truncates them to a maximum length of 512 tokens. Padding ensures that all sequences are of uniform length, and truncation ensures that longer sequences are cut off at the specified length.

````python
# Train-validation split
train_df, val_df = train_test_split(df, test_size=0.8, random_state=42)

# Tokenize the dataset
train_encodings = tokenizer(train_df['Text'].tolist(), truncation=True, padding=True, max_length=512)
val_encodings = tokenizer(val_df['Text'].tolist(), truncation=True, padding=True, max_length=512)
````

This code defines a PyTorch `Dataset` class to handle the tokenized data and labels, and sets up data loaders for efficient batching and iteration during model training and validation.

### Key Components:

1. **EcomDataset Class**:
   - This class is a custom subclass of PyTorch's `Dataset` and is designed to manage the dataset (both inputs and labels) for text classification tasks.
   - **`__init__()`**: Initializes the dataset with `encodings` (tokenized text) and `labels` (numerical label values). The encodings and labels are stored for later access.
   - **`__getitem__()`**: Retrieves a single data sample (input and label) at the specified index `idx`. It creates a dictionary with input IDs, attention masks, and corresponding labels, which are converted to PyTorch tensors.
   - **`__len__()`**: Returns the length of the dataset (number of samples), which is required by PyTorch's `DataLoader`.

2. **Dataset Creation**:
   - The tokenized text data and corresponding labels (from the earlier mapping to `LabelNum`) are used to create `train_dataset` and `val_dataset` objects.
   ```python
   train_dataset = EcomDataset(train_encodings, train_df['LabelNum'].tolist())
   val_dataset = EcomDataset(val_encodings, val_df['LabelNum'].tolist())
   ```

3. **DataLoader**:
   - `DataLoader` is used to wrap the datasets (`train_dataset` and `val_dataset`) to manage batching and shuffling during training and validation.
   - **Batch Size**: Each batch contains 16 samples.
   - **Shuffle**: The training data (`train_loader`) is shuffled before each epoch to ensure that the model does not learn any unintended ordering from the data. The validation data (`val_loader`) is not shuffled to maintain consistency when evaluating model performance.
   ```python
   train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
   val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
   ```

### Purpose:
- **Training**: The `train_loader` will be used to feed batches of tokenized text and their corresponding labels to the model during training.
- **Validation**: The `val_loader` provides the model with validation data, which helps monitor performance and prevent overfitting.

This setup ensures efficient data handling and model training, making it easier to iterate over the dataset in manageable batches.

````python

# Create Dataloader
class EcomDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)


train_dataset = EcomDataset(train_encodings, train_df['LabelNum'].tolist())
val_dataset = EcomDataset(val_encodings, val_df['LabelNum'].tolist())
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
````

This code fine-tunes a pre-trained BERT model on the provided text classification dataset. Below is a breakdown of the main steps:

### Key Components:

1. **Optimizer Initialization**:
   - `AdamW`: The Adam optimizer with weight decay regularization is used. This is typically recommended for fine-tuning transformer models like BERT. The learning rate is set to `2e-5`, which is a common starting point for fine-tuning BERT models.
   ```python
   optimizer = AdamW(model.parameters(), lr=2e-5)
   ```

2. **Training Loop**:
   - The model is fine-tuned for a specified number of epochs (in this case, `num_epochs = 1`), which means the model will go through the entire training dataset once.
   - The loop consists of the following steps:
     - **Set the model to training mode**: `model.train()` ensures that dropout layers (if any) are activated and the model updates during backpropagation.
     - **Initialize the training loss**: `train_loss` is initialized to zero to keep track of the total loss for the epoch.
     - **Batch Processing**:
       - Each batch of data is passed through the model, and the optimizer updates the model's weights accordingly.
       - **Data Transfer to Device**: The input IDs, attention masks, and labels are moved to the specified device (either CPU or GPU).
       - **Forward Pass**: The model processes the input data and outputs the classification predictions. It also computes the loss by comparing predictions with the true labels.
       - **Backpropagation**: The loss is backpropagated through the model, and the optimizer updates the model's parameters based on the gradients.
       - **Loss Accumulation**: The total training loss for the epoch is accumulated.

3. **Loss Calculation**:
   - After processing all batches in the current epoch, the average training loss is printed to monitor the model’s performance during training.
   ```python
   print(f'Epoch {epoch + 1}, Train Loss: {train_loss / len(train_loader)}')
   ```

4. **Completion Message**:
   - After completing the specified number of epochs, a message is printed to indicate that fine-tuning is complete.


This training loop fine-tunes the pre-trained BERT model using the tokenized text data and labels from the training set. The model's parameters are updated via backpropagation, with the optimizer adjusting weights to minimize the loss. The training loss is printed for each epoch to give a sense of how well the model is learning the task.

````python
# Fine-tuning the model
optimizer = AdamW(model.parameters(), lr=2e-5)
num_epochs = 1

for epoch in range(num_epochs):
    model.train()
    train_loss = 0
    for batch in tqdm(train_loader):
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

    print(f'Epoch {epoch + 1}, Train Loss: {train_loss / len(train_loader)}')

print("Fine-tuning complete!")
````

The code snippet defines an evaluation function to assess the model's performance on a validation dataset and calculates key metrics such as **validation loss** and **accuracy**. Here's a breakdown of each part:

### Key Components:

1. **`evaluate` Function**:
   - The function is designed to evaluate a given model using a **dataloader** (which loads batches of data from the validation set). It computes the average loss and accuracy of the model predictions.

2. **Setting the Model to Evaluation Mode**:
   - `model.eval()` ensures the model is in evaluation mode, meaning dropout and other training-specific behaviors are turned off. This is crucial when evaluating to get consistent results.
   
3. **Looping Through the Validation Data**:
   - The function iterates over each batch of validation data:
     - **Data Transfer**: Moves input data (tokenized text and labels) to the device (either CPU or GPU).
     - **Forward Pass**: The model processes the inputs and generates outputs (logits) and computes the loss. Importantly, no gradient calculations are done during evaluation (`torch.no_grad()`), improving efficiency.
   
4. **Loss and Logits Calculation**:
   - The total loss is accumulated (`total_eval_loss`) for each batch. The logits (predicted outputs) are extracted and stored in `predictions`, while the true labels are stored in `true_labels`.

5. **Calculating Metrics**:
   - **Validation Loss**: The average validation loss is calculated by dividing the total loss by the number of batches.
   - **Predictions**: The predicted class labels are obtained by applying `np.argmax()` to the logits (choosing the label with the highest probability).
   - **Validation Accuracy**: The accuracy is computed by comparing the predicted labels (`val_predictions`) with the true labels (`val_true_labels`), and the mean of correct predictions is printed.

### Example of Usage:

```python
# Evaluate the model on the validation set
val_loss, val_predictions, val_true_labels = evaluate(model, val_loader)

# Display the results
print(f"Validation Loss: {val_loss}")
print(f"Validation Accuracy: {(val_predictions == val_true_labels).mean()}")
```

### Output:
- **Validation Loss**: This tells you how well the model is doing on the validation set in terms of how far the predictions are from the true labels.
- **Validation Accuracy**: This metric measures how often the model's predictions match the true labels, providing an overall performance measure.


This evaluation function helps you measure the performance of the fine-tuned model on the validation dataset. By calculating the average loss and accuracy, you can assess how well the model generalizes to unseen data. This is particularly useful in monitoring overfitting or underfitting during the fine-tuning process.

````python
def evaluate(model, dataloader):
    model.eval()
    total_eval_accuracy = 0
    total_eval_loss = 0
    predictions, true_labels = [], []

    for batch in dataloader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        with torch.no_grad():
            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)

        loss = outputs.loss
        logits = outputs.logits
        total_eval_loss += loss.item()

        # Move logits and labels to CPU
        logits = logits.detach().cpu().numpy()
        label_ids = labels.to('cpu').numpy()

        # Store predictions and true labels
        predictions.extend(logits)
        true_labels.extend(label_ids)

    avg_val_loss = total_eval_loss / len(dataloader)
    predictions = np.argmax(predictions, axis=1)

    return avg_val_loss, predictions, true_labels

# Evaluate the model on the validation set
val_loss, val_predictions, val_true_labels = evaluate(model, val_loader)
print(f"Validation Loss: {val_loss}")
print(f"Validation Accuracy: {(val_predictions == val_true_labels).mean()}")
````
